```python
Ans : Nominal encoding is a technique used to transform categorical variables that have no intrinsic ordering 
      into into numerical values that can be used in machine learing models. One common method for nominal 
      encoding os One Hot encoding which creates a binary vector for each catrgory in the variable.
      
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
     "color" : ["red","black","blur","red","black","blue","green","green","red"]
})
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[["color"]])
```

```python
encoded_df = pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
pd.concat([df,encoded_df],axis=1)
```

```python
Q4. Suppose you have a dataset containing categorical data with 5 unique values. Which encoding
    technique would you use to transform this data into a format suitable for machine learning algorithms?
    Explain why you made this choice.
```

```python
Ans: We have a dataset containing categorical data with 5 unique values.
    In the given scenario with a dataset containing categorical data with 5 unique values, I would suggest 
    using one-hot encoding.
    In this technique, each unique category is converted into a binary vector, where each element represents 
    the presence or absence of a particular category. For example, if we have five unique categories, we would
    create five binary features.
    Suppose we have a categorical feature "Color" with five unique categories: Red, Blue, Green, Yellow, 
    and Purple. After one-hot encoding, we would create five binary features: Color_Red, Color_Blue, 
    Color_Green, Color_Yellow, and Color_Purple. Each feature would have a value of 1 if it corresponds 
    to the respective category and 0 otherwise.
    The reason why one-hot encoding is a suitable choice in this scenario is that it effectively represents
    categorical data in a format that machine learning algorithms can interpret. One-hot encoding ensures that
    there is no inherent ordinal relationship between the categories.
```

```python
Q6. You are working with a dataset containing information about different types of animals, including their
   species, habitat, and diet. Which encoding technique would you use to transform the categorical data into
   a format suitable for machine learning algorithms? Justify your answer.
```

```python
Ans: The choice between any of these encoding techniques depends on the characteristics of the categorical 
     variables and the specific requirements of the machine learning task.
    
    If the categorical variables have an inherent  ordinal relationship, label encoding may be appropriate.
    For example, in the case of species, we could assign labels like 1 for "Dog," 2 for "Cat," 3 for "Bird," 
    and so on. 
    
    If there is no ordinal relationship and each category is independent, one-hot encoding would be a suitable
    choice. This technique is useful when there is no ordinal relationship between the categories, which is 
    often the case with nominal variables like species, habitat, and diet.
    
    Target encoding can be useful if there is a strong relationship between the categorical variable and the 
    target variable.
    For example, if we have a target variable like "animal weight," target encoding would replace each species
    category with the average weight of animals belonging to that species.
```
