```python
Ans -: ArithmeticError is simply an error that occurs during numeric calculations.
        Error which are occur do mathmetical computation in the programe are comes under the ArithmeticError.
```

```python
# ZeroDivisionError
try:
    5/0
except ZeroDivisionError as e:
    print(e)
```

```python
This showes the ZeroDivisionError which occur the a number is under goes division of 0.
```

```python
# OverflowError

j = 5.0
try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("Overflow error happened")
    print(f"{e}, {e.__class__}")
```

```python
An OverflowError exception is raised when an arithmetic operation exceeds the limits to be represented.
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
Ans -:  The LookupError exception in Python forms the base class for all exceptions 
        that are raised when an index or a key is not found for a sequence or dictionary respectively
```

```python
# IndentationError
li = [1,2,3,4,5]
try:
    print(li[10])
except LookupError as e:
    print(e)
```

```python
# KeyError

d = {"k1":1,"k2":3}
try:
     c =  d["k4"] 
except LookupError as e:
    print(e,e.__class__)
```
