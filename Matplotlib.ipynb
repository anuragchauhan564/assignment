```python
Q2 : What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
      plot a scatter plot.
        
        import numpy as np
        np.random.seed(3)
        x = 3 + np.random.normal(0,2,50)
        y = 3 + np.random.normal(0,2,len(x))
```

```python
Ans : A Scatter plot is a diagram where each value in the data set is represented by a dot.
       Scatter plot needs two array of the same lenght, one for the values of the x- axis and one for 
        the values of the y- axis.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(x,y)
```

```python
Q3 : Why is the subplot() function used? Draw four line plots using the subplot()
     function.
```

```python
Ans :  Subplot() method provides a way to plot multiple plots on a single figure. 
       The subplt() function takes three arguments that describes the layout of the figure.
        The layout is organized in rows and columns which are represented by the first and second 
        argument.
        the third argument represents the index of the current plot.
```

```python
For line 1 : x = np.array([0,1,2,3,4,5]) and y = np.array([0,100,200,300,400,500])
For line 1 : x = np.array([0,1,2,3,4,5]) and y = np.array([50,20,40,20,60,70])
For line 1 : x = np.array([0,1,2,3,4,5]) and y = np.array([10,20,30,40,50,60])
For line 1 : x = np.array([0,1,2,3,4,5]) and y = np.array([200,350,250,550,450,150])
```

```python
# For line 1 :
x = np.array([0,1,2,3,4,5]) 
y = np.array([0,100,200,300,400,500])
plt.subplot(2,2,1)
plt.plot(x,y)

# For line 1 :
x = np.array([0,1,2,3,4,5]) 
y = np.array([50,20,40,20,60,70])
plt.subplot(2,2,2)
plt.plot(x,y)

# For line 1 :
x = np.array([0,1,2,3,4,5])
y = np.array([10,20,30,40,50,60])
plt.subplot(2,2,3)
plt.plot(x,y)

# For line 1 :
x = np.array([0,1,2,3,4,5])
y = np.array([200,350,250,550,450,150])
plt.subplot(2,2,4)
plt.plot(x,y)

plt.show()
```

```python
Q4 : What is a bar plot? Why is it used? Using the following data plot a bar plot and a 
    bar plot and a horizontal bar plot.
```

```python
Ans : Bar plots are a type of data visualization used to represent data in the form of 
      rectangular bars. The height of each bar represent the value of a data point, and 
      the width of each bar represents the category of the data.
    
    Bar plot is used to create bar plots.
```

```python
company = np.array(["Apple","Microsoft","Google","AMD"])
profit = np.array([3000,8000,1000,10000])
```

```python
plt.bar(company,profit)
plt.title("Bar Plot")
plt.ylabel("Profit")
plt.xlabel("Company")
```

```python
Q5 : What is a box plot? Why is it used? Using the following data plot a box plot.
    box1 = np.random.normal(100,10,200)
    box2 = np.random.normal(90,20,200)
```

```python
Ans :  Boxplot are a measure of how well distributed the data in a data set is.
        It divides the data set into three quartiles.This graph represent the 
        minimum, maximum, median, first quartile and thrid quartile in the data set.
        
        It is used to show distributions of numeric data values, especiall when you 
        want to compare them between multiple groups.
```

```python
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
data = [box1,box2]
fig = plt.figure(figsize=(3,3))
ax = fig.add_axes([0,0,1,1])
bp =ax.boxplot(data)
plt.show()
```
